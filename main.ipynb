```python
import sys
import importlib
from pathlib import Path

# Ensure current directory is in path
sys.path.insert(0, str(Path.cwd()))

# Remove any previously imported etl_py modules to force reload
for key in list(sys.modules.keys()):
    if 'etl_py' in key:
        del sys.modules[key]

from etl_py.extract import extract_data
from etl_py.transform import transform_data
from etl_py.load import load_data
```

```python
def run_etl():

    # 1. Extract
    df_bookings, df_grounds, df_users = extract_data()
    print(" =========== Extract Completed:========")

    # 2. Transform
    dim_users, dim_grounds, dim_date, fact_bookings = transform_data(df_bookings, df_grounds, df_users)
    print(" =========== Transform Completed:========")

    # 3. Load
    load_data(dim_users, dim_grounds, dim_date, fact_bookings,schema='public')
    print(" =========== Load Completed:========")
```

```python
if __name__ == "__main__":
    run_etl()
    print(" ============== THANK YOU =============");
```
